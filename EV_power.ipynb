```python
# ============================================================
# EV Smart Power Management — Option 3: Multi-Output Regression
# Phase 1: NEV Dataset (Development)
# Phase 2: Real Driving Dataset (Validation)
# ============================================================

# Run this first in a separate cell:
# !pip install imbalanced-learn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')

from google.colab import drive
drive.mount('/content/drive')

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import joblib

# ============================================================
# HELPER FUNCTIONS
# ============================================================

def assign_p1(row):
    """Propulsion power allocation"""
    soc = row['SOC']
    if soc < 20:
        return 0.70
    elif soc < 30:
        return 0.85
    else:
        return 1.00

def assign_p2(row):
    """Auxiliary power allocation (AC, heating)"""
    soc   = row['SOC']
    power = row['PowerDemand']
    regen = row['RegenBraking']
    if soc < 20 or power > 70:
        return 0.00
    elif soc < 30 or power > 60:
        return 0.20
    elif soc < 50 or power > 50:
        return 0.50
    elif regen > 3:
        return 1.00
    else:
        return 1.00

def assign_p3(row):
    """Infotainment power allocation"""
    soc   = row['SOC']
    power = row['PowerDemand']
    regen = row['RegenBraking']
    if soc < 30 or power > 60:
        return 0.00
    elif soc < 50 or power > 50:
        return 0.30
    elif regen > 3:
        return 1.00
    else:
        return 1.00

def evaluate_model(y_true, y_pred, output_names):
    """Evaluate multi-output regression results"""
    print(f"\n{'Output':<20} {'MAE':>8} {'RMSE':>8} {'R²':>8}")
    print("-" * 46)
    results = {}
    for i, name in enumerate(output_names):
        mae  = mean_absolute_error(y_true[:, i], y_pred[:, i])
        rmse = np.sqrt(mean_squared_error(y_true[:, i], y_pred[:, i]))
        r2   = r2_score(y_true[:, i], y_pred[:, i])
        print(f"{name:<20} {mae:>8.4f} {rmse:>8.4f} {r2:>8.4f}")
        results[name] = {'MAE': mae, 'RMSE': rmse, 'R2': r2}
    return results

def plot_actual_vs_predicted(y_true, y_pred, output_names, title, savepath):
    """Plot actual vs predicted for each output"""
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))
    colors = ['#e74c3c', '#3498db', '#2ecc71']
    for i, (name, ax, color) in enumerate(zip(output_names, axes, colors)):
        ax.plot(y_true[:200, i], label='Actual', color='black', linewidth=1)
        ax.plot(y_pred[:200, i], label='Predicted', color=color,
                linestyle='--', linewidth=1)
        ax.set_title(f'{name}')
        ax.set_xlabel('Timestep')
        ax.set_ylabel('Power Allocation (0-1)')
        ax.legend()
        ax.set_ylim(-0.1, 1.2)
    plt.suptitle(title, fontsize=13, fontweight='bold')
    plt.tight_layout()
    plt.savefig(savepath, dpi=150)
    plt.show()

def plot_power_allocation_over_time(y_pred, output_names, title, savepath):
    """Stacked area chart of power allocation over time"""
    plt.figure(figsize=(14, 5))
    timesteps = np.arange(min(300, len(y_pred)))
    plt.fill_between(timesteps, y_pred[:300, 0],
                     alpha=0.6, color='#e74c3c', label='P1 Propulsion')
    plt.fill_between(timesteps, y_pred[:300, 1],
                     alpha=0.6, color='#3498db', label='P2 Auxiliary')
    plt.fill_between(timesteps, y_pred[:300, 2],
                     alpha=0.6, color='#2ecc71', label='P3 Infotainment')
    plt.title(title)
    plt.xlabel('Timestep')
    plt.ylabel('Power Allocation (0 = OFF, 1 = FULL)')
    plt.legend()
    plt.tight_layout()
    plt.savefig(savepath, dpi=150)
    plt.show()

def plot_feature_importance(model, feature_names, output_names, savepath):
    """Feature importance for each output"""
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    for i, (name, ax) in enumerate(zip(output_names, axes)):
        imp = pd.Series(
            model.estimators_[i].feature_importances_,
            index=feature_names
        ).sort_values()
        imp.plot(kind='barh', ax=ax, color='steelblue')
        ax.set_title(f'Feature Importance — {name}')
        ax.set_xlabel('Importance Score')
    plt.tight_layout()
    plt.savefig(savepath, dpi=150)
    plt.show()

def plot_error_distribution(y_true, y_pred, output_names, title, savepath):
    """Error distribution histogram for each output"""
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))
    colors = ['#e74c3c', '#3498db', '#2ecc71']
    for i, (name, ax, color) in enumerate(zip(output_names, axes, colors)):
        errors = y_pred[:, i] - y_true[:, i]
        ax.hist(errors, bins=40, color=color, edgecolor='black', alpha=0.7)
        ax.axvline(0, color='black', linestyle='--', linewidth=1)
        ax.set_title(f'Error Distribution — {name}')
        ax.set_xlabel('Prediction Error')
        ax.set_ylabel('Frequency')
    plt.suptitle(title, fontsize=13, fontweight='bold')
    plt.tight_layout()
    plt.savefig(savepath, dpi=150)
    plt.show()

OUTPUT_NAMES = [
    'P1_Propulsion', 'P2_Auxiliary',
    'P3_Infotainment', 'RemainingRange_km'
]
DRIVE_PATH   = '/content/drive/MyDrive/final yr'

# ============================================================
# PHASE 1 — NEV DATASET (Development)
# ============================================================

print("=" * 60)
print("PHASE 1 — NEV DATASET")
print("=" * 60)

# ── Load ────────────────────────────────────────────────────
nev = pd.read_csv(f'{DRIVE_PATH}/nev_energy_management_dataset.csv',
                  encoding='latin1', on_bad_lines='skip')

# ── Preprocess ──────────────────────────────────────────────
nev.dropna(inplace=True)
nev.drop_duplicates(inplace=True)
nev = nev.sort_values('Time (s)').reset_index(drop=True)

# Outlier removal
num_cols = ['Speed (km/h)', 'Acceleration (m/sÂ²)', 'Power Demand (kW)',
            'Battery Power (kW)', 'SOC (%)',
            'Regenerative Braking Power (kW)']
for col in num_cols:
    Q1, Q3 = nev[col].quantile(0.25), nev[col].quantile(0.75)
    IQR = Q3 - Q1
    nev = nev[(nev[col] >= Q1-1.5*IQR) & (nev[col] <= Q3+1.5*IQR)]

nev = nev.reset_index(drop=True)

# Drop highly correlated feature
nev.drop(columns=['Battery Degradation (%)'], inplace=True)

# Encode driving cycle
nev['Driving_Cycle_Encoded'] = nev['Driving Cycle Type'].map({
    'Urban': 0, 'Highway': 1, 'WLTC': 2, 'HWFET': 3
})

# ── Create standardized columns for label functions ─────────
nev['SOC']          = nev['SOC (%)']
nev['PowerDemand']  = nev['Power Demand (kW)']
nev['RegenBraking'] = nev['Regenerative Braking Power (kW)']

# ── Create P1, P2, P3 targets ───────────────────────────────
nev['P1'] = nev.apply(assign_p1, axis=1)
nev['P2'] = nev.apply(assign_p2, axis=1)
nev['P3'] = nev.apply(assign_p3, axis=1)

print("\n=== NEV Power Allocation Distribution ===")
print(f"P1 — mean: {nev['P1'].mean():.3f}, unique: {sorted(nev['P1'].unique())}")
print(f"P2 — mean: {nev['P2'].mean():.3f}, unique: {sorted(nev['P2'].unique())}")
print(f"P3 — mean: {nev['P3'].mean():.3f}, unique: {sorted(nev['P3'].unique())}")

# ── Features and Targets ────────────────────────────────────
nev_feature_cols = [
    'Speed (km/h)', 'Acceleration (m/sÂ²)', 'SOC (%)',
    'Battery Power (kW)', 'Power Demand (kW)',
    'Regenerative Braking Power (kW)', 'Driving_Cycle_Encoded'
]

X_nev = nev[nev_feature_cols].values
nev['RemainingRange'] = (nev['SOC (%)'] / 100) * 150
y_nev = nev[['P1', 'P2', 'P3', 'RemainingRange']].values

# ── Normalize ───────────────────────────────────────────────
nev_scaler = MinMaxScaler()
X_nev_scaled = nev_scaler.fit_transform(X_nev)
joblib.dump(nev_scaler, f'{DRIVE_PATH}/nev_scaler.pkl')

# ── Train/Test Split ────────────────────────────────────────
X_train_nev, X_test_nev, y_train_nev, y_test_nev = train_test_split(
    X_nev_scaled, y_nev, test_size=0.2, random_state=42
)
print(f"\nNEV — Train: {X_train_nev.shape[0]}, Test: {X_test_nev.shape[0]}")

# ── Train Multi-Output Random Forest ────────────────────────
print("\nTraining Multi-Output Random Forest on NEV data...")
nev_model = MultiOutputRegressor(
    RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)
)
nev_model.fit(X_train_nev, y_train_nev)

# ── Evaluate ────────────────────────────────────────────────
y_pred_nev = nev_model.predict(X_test_nev)

print("\n=== NEV Model Results ===")
nev_results = evaluate_model(y_test_nev, y_pred_nev, OUTPUT_NAMES)

# ── Plots ───────────────────────────────────────────────────
plot_actual_vs_predicted(
    y_test_nev, y_pred_nev, OUTPUT_NAMES,
    'NEV Dataset — Actual vs Predicted Power Allocation',
    f'{DRIVE_PATH}/nev_actual_vs_predicted.png'
)

plot_power_allocation_over_time(
    y_pred_nev, OUTPUT_NAMES,
    'NEV Dataset — Predicted Power Allocation Over Time',
    f'{DRIVE_PATH}/nev_power_allocation.png'
)

plot_feature_importance(
    nev_model, nev_feature_cols, OUTPUT_NAMES,
    f'{DRIVE_PATH}/nev_feature_importance.png'
)

plot_error_distribution(
    y_test_nev, y_pred_nev, OUTPUT_NAMES,
    'NEV Dataset — Prediction Error Distribution',
    f'{DRIVE_PATH}/nev_error_distribution.png'
)

# ── Save NEV model ───────────────────────────────────────────
joblib.dump(nev_model, f'{DRIVE_PATH}/nev_multioutput_model.pkl')
print("\n✅ NEV model saved.")

# ============================================================
# PHASE 2 — REAL DRIVING DATASET (Validation)
# ============================================================

print("\n" + "=" * 60)
print("PHASE 2 — REAL DRIVING DATASET")
print("=" * 60)

# ── Load and merge all trip files ───────────────────────────
real_folder = f'{DRIVE_PATH}/real_driving_data'
trip_files  = sorted([f for f in os.listdir(real_folder)
                      if f.endswith('.csv')])

print(f"Found {len(trip_files)} trip files. Loading...")

dfs = []
for i, fname in enumerate(trip_files):
    try:
        trip = pd.read_csv(
            f"{real_folder}/{fname}",
            encoding='latin1',
            sep=None,
            engine='python',
            on_bad_lines='skip'
        )
        trip['Trip_ID'] = i  # tag which trip each row came from
        dfs.append(trip)
    except Exception as e:
        print(f"Skipped {fname}: {e}")

real = pd.concat(dfs, ignore_index=True)
print(f"Total rows after merging: {real.shape[0]}")

# ── Compute Battery Power from Voltage × Current ────────────
real['Battery_Power_kW'] = (real['Battery Voltage [V]'] *
                             real['Battery Current [A]']) / 1000.0

# ── Compute Power Demand from Throttle × Torque ─────────────
# Normalize to kW range similar to NEV dataset
real['Power_Demand_kW'] = (real['Throttle [%]'] *
                            real['Motor Torque [Nm]'].abs()) / 100.0

# ── Select and rename relevant columns ──────────────────────
real = real[[
    'Time [s]', 'Velocity [km/h]', 'Longitudinal Acceleration [m/s^2]',
    'SoC [%]', 'Battery_Power_kW', 'Power_Demand_kW',
    'Regenerative Braking Signal ', 'AirCon Power [kW]',
    'Heating Power CAN [kW]',
    'Ambient Temperature [°C]',        # ADD THIS
    'Cabin Temperature Sensor [°C]',   # ADD THIS
    'Trip_ID'
]].copy()

real.columns = [
    'Time', 'Speed', 'Acceleration', 'SOC',
    'BatteryPower', 'PowerDemand', 'RegenBraking',
    'AirCon_kW', 'Heating_kW',
    'AmbientTemp',    # ADD THIS
    'CabinTemp',      # ADD THIS
    'Trip_ID'
]

# ── Clean ───────────────────────────────────────────────────
real.dropna(inplace=True)
real.drop_duplicates(inplace=True)

# Outlier removal per trip to preserve continuity
real_clean = []
for tid in real['Trip_ID'].unique():
    trip_df = real[real['Trip_ID'] == tid].copy()
    for col in ['Speed', 'Acceleration', 'SOC',
                'BatteryPower', 'PowerDemand']:
        Q1, Q3 = trip_df[col].quantile(0.25), trip_df[col].quantile(0.75)
        IQR = Q3 - Q1
        trip_df = trip_df[
            (trip_df[col] >= Q1-1.5*IQR) &
            (trip_df[col] <= Q3+1.5*IQR)
        ]
    real_clean.append(trip_df)

real = pd.concat(real_clean, ignore_index=True)
print(f"Real data after cleaning: {real.shape[0]} rows")

# ── SOC distribution in real data ───────────────────────────
plt.figure(figsize=(7, 4))
plt.hist(real['SOC'], bins=40, color='steelblue', edgecolor='black')
plt.title('Real Driving Data — SOC Distribution')
plt.xlabel('SOC (%)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig(f'{DRIVE_PATH}/real_soc_distribution.png', dpi=150)
plt.show()

# ── Create P1 using same logic as NEV ───────────────────────
real['P1'] = real.apply(assign_p1, axis=1)

# ── For P2 and P3: normalize real measurements to 0-1 ───────
# AirCon Power → P2 (normalize by max value)
aircon_max  = real['AirCon_kW'].max()
heating_max = real['Heating_kW'].max()

real['P2'] = (real['AirCon_kW'] / aircon_max).clip(0, 1) \
             if aircon_max > 0 else real.apply(assign_p2, axis=1)
real['P3'] = (real['Heating_kW'] / heating_max).clip(0, 1) \
             if heating_max > 0 else real.apply(assign_p3, axis=1)

print("\n=== Real Data Power Allocation Distribution ===")
print(f"P1 — mean: {real['P1'].mean():.3f}")
print(f"P2 — mean: {real['P2'].mean():.3f} (from real AirCon measurements)")
print(f"P3 — mean: {real['P3'].mean():.3f} (from real Heating measurements)")

# ── Features and Targets ────────────────────────────────────
real_feature_cols = [
    'Speed', 'Acceleration', 'SOC',
    'BatteryPower', 'PowerDemand', 'RegenBraking',
    'AmbientTemp',   # ADD THIS
    'CabinTemp'      # ADD THIS
]

X_real = real[real_feature_cols].values
real['RemainingRange'] = (real['SOC'] / 100) * 150
y_real = real[['P1', 'P2', 'P3', 'RemainingRange']].values

# ── Normalize ───────────────────────────────────────────────
real_scaler = MinMaxScaler()
X_real_scaled = real_scaler.fit_transform(X_real)
joblib.dump(real_scaler, f'{DRIVE_PATH}/real_scaler.pkl')

# ── Train/Test Split ────────────────────────────────────────
X_train_real, X_test_real, y_train_real, y_test_real = train_test_split(
    X_real_scaled, y_real, test_size=0.2, random_state=42
)
print(f"\nReal — Train: {X_train_real.shape[0]}, Test: {X_test_real.shape[0]}")

# ── Train Multi-Output Random Forest on Real Data ───────────
print("\nTraining Multi-Output Random Forest on real driving data...")
real_model = MultiOutputRegressor(
    RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)
)
real_model.fit(X_train_real, y_train_real)

# ── Evaluate ────────────────────────────────────────────────
y_pred_real = real_model.predict(X_test_real)

print("\n=== Real Driving Dataset Results ===")
real_results = evaluate_model(y_test_real, y_pred_real, OUTPUT_NAMES)

# ── Plots ───────────────────────────────────────────────────
plot_actual_vs_predicted(
    y_test_real, y_pred_real, OUTPUT_NAMES,
    'Real Driving Data — Actual vs Predicted Power Allocation',
    f'{DRIVE_PATH}/real_actual_vs_predicted.png'
)

plot_power_allocation_over_time(
    y_pred_real, OUTPUT_NAMES,
    'Real Driving Data — Predicted Power Allocation Over Time',
    f'{DRIVE_PATH}/real_power_allocation.png'
)

plot_feature_importance(
    real_model, real_feature_cols, OUTPUT_NAMES,
    f'{DRIVE_PATH}/real_feature_importance.png'
)

plot_error_distribution(
    y_test_real, y_pred_real, OUTPUT_NAMES,
    'Real Driving Data — Prediction Error Distribution',
    f'{DRIVE_PATH}/real_error_distribution.png'
)

# ── Save real model ──────────────────────────────────────────
joblib.dump(real_model, f'{DRIVE_PATH}/real_multioutput_model.pkl')
print("\n✅ Real driving model saved.")

# ============================================================
# PHASE 3 — COMPARISON: NEV vs REAL DRIVING
# ============================================================

print("\n" + "=" * 60)
print("PHASE 3 — COMPARISON: NEV vs REAL DRIVING")
print("=" * 60)

metrics = ['MAE', 'RMSE', 'R2']
fig, axes = plt.subplots(1, 4, figsize=(20, 5))  # changed 3 to 4

for i, output in enumerate(OUTPUT_NAMES):
    nev_vals  = [nev_results[output][m]  for m in metrics]
    real_vals = [real_results[output][m] for m in metrics]
    x = np.arange(len(metrics))
    axes[i].bar(x - 0.2, nev_vals,  0.4, label='NEV',
                color='steelblue', alpha=0.8)
    axes[i].bar(x + 0.2, real_vals, 0.4, label='Real Driving',
                color='darkorange', alpha=0.8)
    axes[i].set_title(f'{output}')
    axes[i].set_xticks(x)
    axes[i].set_xticklabels(metrics)
    axes[i].legend()
    axes[i].set_ylabel('Score')

plt.suptitle('NEV vs Real Driving — Model Performance Comparison',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig(f'{DRIVE_PATH}/comparison_nev_vs_real.png', dpi=150)
plt.show()

# ============================================================
# PHASE 4 — END-TO-END PIPELINE DEMONSTRATION
# ============================================================

print("\n" + "=" * 60)
print("END-TO-END PIPELINE DEMONSTRATION")
print("=" * 60)

# Sample input — simulate a real-time sensor reading
sample_input = {
    'Speed':        72.0,
    'Acceleration': -0.3,
    'SOC':          28.0,
    'BatteryPower': 18.5,
    'PowerDemand':  55.0,
    'RegenBraking':  0.5,
    'AmbientTemp':  28.0,   # ADD THIS — hot day in Kerala
    'CabinTemp':    24.0    # ADD THIS — target cabin comfort temp
}

print("\nSensor Input:")
for k, v in sample_input.items():
    print(f"  {k:<20}: {v}")

# Scale input
X_demo = real_scaler.transform([[
    sample_input['Speed'],
    sample_input['Acceleration'],
    sample_input['SOC'],
    sample_input['BatteryPower'],
    sample_input['PowerDemand'],
    sample_input['RegenBraking'],
    sample_input['AmbientTemp'],   # ADD THIS
    sample_input['CabinTemp']      # ADD THIS
]])

# Predict
p1, p2, p3, remaining_range = real_model.predict(X_demo)[0]
p1, p2, p3 = np.clip([p1, p2, p3], 0, 1)
remaining_range = np.clip(remaining_range, 0, 150)

print(f"\nML Model Output:")
print(f"  P1 Propulsion    : {p1*100:.1f}%")
print(f"  P2 Auxiliary     : {p2*100:.1f}%")
print(f"  P3 Infotainment  : {p3*100:.1f}%")
print(f"  Remaining Range  : {remaining_range:.1f} km")

print(f"\nML Model Output:")
print(f"  P1 Propulsion  : {p1*100:.1f}%")
print(f"  P2 Auxiliary   : {p2*100:.1f}%")
print(f"  P3 Infotainment: {p3*100:.1f}%")

# Quantize to 2-bit for FSM
def quantize(val):
    if val < 0.25:   return '00'  # OFF
    elif val < 0.50: return '01'  # 25%
    elif val < 0.75: return '10'  # 50%
    else:            return '11'  # FULL

p1_bits = quantize(p1)
p2_bits = quantize(p2)
p3_bits = quantize(p3)

print(f"\nFSM 2-bit Control Signals:")
print(f"  P1 → {p1_bits} (Propulsion)")
print(f"  P2 → {p2_bits} (Auxiliary)")
print(f"  P3 → {p3_bits} (Infotainment)")
print(f"\nCombined 6-bit FSM input: {p1_bits}{p2_bits}{p3_bits}")

# ── Power allocation bar chart for this sample ──────────────
plt.figure(figsize=(6, 4))
bars = plt.bar(['P1\nPropulsion', 'P2\nAuxiliary', 'P3\nInfotainment'],
               [p1, p2, p3],
               color=['#e74c3c', '#3498db', '#2ecc71'],
               edgecolor='black')
for bar, val in zip(bars, [p1, p2, p3]):
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 0.02,
             f'{val*100:.1f}%', ha='center', fontweight='bold')
plt.ylim(0, 1.2)
plt.title(f'Power Allocation — SOC={sample_input["SOC"]}%, '
          f'Demand={sample_input["PowerDemand"]}kW')
plt.ylabel('Power Allocation (0=OFF, 1=FULL)')
plt.tight_layout()
plt.savefig(f'{DRIVE_PATH}/demo_power_allocation.png', dpi=150)
plt.show()



# ============================================================
# BASELINE COMPARISON — Rule-Based BMS vs ML Model
# ============================================================

print("\n" + "="*55)
print("BASELINE COMPARISON — Rule-Based BMS vs ML Model")
print("="*55)

def rule_based_bms(soc):
    """Traditional BMS — only looks at SOC, ignores everything else"""
    if soc < 20:
        return [0.70, 0.00, 0.00]
    elif soc < 30:
        return [0.85, 0.20, 0.00]
    elif soc < 50:
        return [1.00, 0.50, 0.30]
    else:
        return [1.00, 1.00, 1.00]

# Get original unscaled SOC for test samples
# We use the last 20% of real data as test (same split)
soc_test_values = real['SOC'].values[-len(X_test_real):]

baseline_preds = np.array([
    rule_based_bms(soc) for soc in soc_test_values
])


baseline_output_names = ['P1_Propulsion', 'P2_Auxiliary', 'P3_Infotainment']

# Compare
print(f"\n{'Output':<20} {'Rule-Based MAE':>15} "
      f"{'ML Model MAE':>13} {'Improvement':>12}")
print("-"*62)

for i, name in enumerate(baseline_output_names):
    baseline_mae = mean_absolute_error(
        y_test_real[:, i], baseline_preds[:, i])
    ml_mae = mean_absolute_error(
        y_test_real[:, i], y_pred_real[:, i])
    improvement = ((baseline_mae - ml_mae) /
                   (baseline_mae + 1e-10)) * 100
    print(f"{name:<20} {baseline_mae:>15.4f} "
          f"{ml_mae:>13.4f} {improvement:>11.1f}%")

# Plot — only P1, P2, P3 (not RemainingRange)
baseline_output_names = ['P1_Propulsion', 'P2_Auxiliary', 'P3_Infotainment']
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
colors = ['#e74c3c', '#3498db', '#2ecc71']

for i, (name, ax, color) in enumerate(
        zip(baseline_output_names, axes, colors)):
    bl_mae = mean_absolute_error(
        y_test_real[:, i], baseline_preds[:, i])
    ml_mae = mean_absolute_error(
        y_test_real[:, i], y_pred_real[:, i])
    improvement = ((bl_mae - ml_mae) / (bl_mae + 1e-10)) * 100

    ax.bar(['Rule-Based\nBMS', 'ML Model\n(Ours)'],
           [bl_mae, ml_mae],
           color=['#95a5a6', color],
           edgecolor='black', width=0.4)
    ax.set_title(f'{name}')
    ax.set_ylabel('MAE (Lower is Better)')
    ax.text(0.5, max(bl_mae, ml_mae) * 0.6,
            f'{improvement:.1f}% better',
            ha='center', fontsize=11,
            fontweight='bold', color='green')

plt.suptitle(
    'Rule-Based BMS vs ML Model — MAE Comparison',
    fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig(f'{DRIVE_PATH}/baseline_vs_ml_comparison.png', dpi=150)
plt.show()
print("✅ Baseline comparison saved.")




# ============================================================
# SINGLE TRIP VALIDATION — TripA01 (Real 16-min BMW i3 Drive)
# ============================================================

print("\n" + "="*55)
print("SINGLE TRIP VALIDATION — TripA01")
print("="*55)

# Train on all trips except TripA01, test on TripA01
trip_test  = real[real['Trip_ID'] == 0].copy()
trip_train = real[real['Trip_ID'] != 0].copy()

print(f"TripA01 test rows : {len(trip_test)}")
print(f"Training rows     : {len(trip_train)}")

X_trip_train = trip_train[real_feature_cols].values
y_trip_train = trip_train[['P1', 'P2', 'P3']].values
X_trip_test  = trip_test[real_feature_cols].values
y_trip_test  = trip_test[['P1', 'P2', 'P3']].values

trip_scaler = MinMaxScaler()
X_trip_train_sc = trip_scaler.fit_transform(X_trip_train)
X_trip_test_sc  = trip_scaler.transform(X_trip_test)

trip_model = MultiOutputRegressor(
    RandomForestRegressor(
        n_estimators=100, max_depth=10, random_state=42)
)
trip_model.fit(X_trip_train_sc, y_trip_train)
y_trip_pred = trip_model.predict(X_trip_test_sc)

print("\n=== TripA01 Holdout Test Results ===")
for i, name in enumerate(baseline_output_names):
    mae = mean_absolute_error(y_trip_test[:, i], y_trip_pred[:, i])
    r2  = r2_score(y_trip_test[:, i], y_trip_pred[:, i])
    print(f"{name:<20} MAE: {mae:.4f}   R²: {r2:.4f}")

# Plot full trip — SOC discharge + power allocation over time
time_min = trip_test['Time'].values / 60

fig, axes = plt.subplots(4, 1, figsize=(14, 12), sharex=True)

# SOC
axes[0].plot(time_min, trip_test['SOC'].values,
             color='orange', linewidth=1.5)
axes[0].axhline(y=30, color='red', linestyle='--',
                alpha=0.6, label='Critical (30%)')
axes[0].axhline(y=50, color='gold', linestyle='--',
                alpha=0.6, label='Moderate (50%)')
axes[0].set_ylabel('SOC (%)')
axes[0].set_title('Battery State of Charge During Trip')
axes[0].legend(fontsize=8)

# P1
axes[1].plot(time_min, y_trip_pred[:, 0],
             color='#e74c3c', linewidth=1)
axes[1].fill_between(time_min, y_trip_pred[:, 0],
                     alpha=0.3, color='#e74c3c')
axes[1].set_ylabel('P1 Propulsion')
axes[1].set_ylim(0, 1.1)
axes[1].set_title('Propulsion Power Allocation')

# P2
axes[2].plot(time_min, y_trip_pred[:, 1],
             color='#3498db', linewidth=1)
axes[2].fill_between(time_min, y_trip_pred[:, 1],
                     alpha=0.3, color='#3498db')
axes[2].set_ylabel('P2 Auxiliary')
axes[2].set_ylim(0, 1.1)
axes[2].set_title('Auxiliary Power Allocation (AC/Heating)')

# P3
axes[3].plot(time_min, y_trip_pred[:, 2],
             color='#2ecc71', linewidth=1)
axes[3].fill_between(time_min, y_trip_pred[:, 2],
                     alpha=0.3, color='#2ecc71')
axes[3].set_ylabel('P3 Infotainment')
axes[3].set_xlabel('Time (minutes)')
axes[3].set_ylim(0, 1.1)
axes[3].set_title('Infotainment Power Allocation')

plt.suptitle(
    'TripA01 — Real BMW i3 Drive: '
    'Intelligent Power Management Over Time',
    fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig(f'{DRIVE_PATH}/trip_validation_power_over_time.png',
            dpi=150)
plt.show()
print("✅ Single trip validation saved.")





# ============================================================
# FINAL SUMMARY
# ============================================================

print("\n" + "=" * 60)
print("FINAL RESULTS SUMMARY")
print("=" * 60)
print(f"\n{'Dataset':<15} {'Output':<20} {'MAE':>8} {'RMSE':>8} {'R²':>8}")
print("-" * 63)
for output in OUTPUT_NAMES:
    n = nev_results[output]
    r = real_results[output]
    print(f"{'NEV':<15} {output:<20} "
          f"{n['MAE']:>8.4f} {n['RMSE']:>8.4f} {n['R2']:>8.4f}")
    print(f"{'Real Driving':<15} {output:<20} "
          f"{r['MAE']:>8.4f} {r['RMSE']:>8.4f} {r['R2']:>8.4f}")
    print()

print("✅ All models and plots saved to Google Drive → final yr folder")
print("✅ ML phase complete. Ready for Verilog FSM implementation.")
```
